```python
#BÀI TOÁN DỰ BÁO THỜI TIẾT
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
wind=ctrl.Antecedent(np.arange(0.7,3.1,0.1),'wind')
temp=ctrl.Antecedent(np.arange(20.4, 41.2,0.1),'temp')
rain= ctrl.Consequent(np.arange(0.2,9.1,0.1),'rain')

wind['veryhigh']=fuzz.trimf(wind.universe,[0.7,0.7,1.5])
wind['high']=fuzz.trimf(wind.universe,[0.7,1.5,2])
wind['normal']=fuzz.trimf(wind.universe,[1.5,2,2.8])
wind['low']=fuzz.trimf(wind.universe,[2,2.8,3.1])
wind['verylow']=fuzz.trimf(wind.universe,[2.8,3.1,3.1])

temp['veryhigh']=fuzz.trimf(temp.universe,[20.4,20.4,25])
temp['high']=fuzz.trimf(temp.universe,[20.4,25,30])
temp['normal']=fuzz.trimf(temp.universe,[25,30,35])
temp['low']=fuzz.trimf(temp.universe,[30,35,41.2])
temp['verylow']=fuzz.trimf(temp.universe,[35,41.2,41.2])

rain['veryhigh']=fuzz.trimf(rain.universe,[0.2,1,2.5])
rain['high']=fuzz.trimf(rain.universe,[1,2.5,4.5])
rain['normal']=fuzz.trimf(rain.universe,[2.5,4.5,6.5])
rain['low']=fuzz.trimf(rain.universe,[4.5,6.5,8])
rain['verylow']=fuzz.trimf(rain.universe,[6.5,8,9.1])

rule1=ctrl.Rule(wind['veryhigh'] & temp['veryhigh'],rain['veryhigh'])
rule2=ctrl.Rule(wind['veryhigh'] & temp['high'],rain['high'])
rule3=ctrl.Rule(wind['veryhigh'] & temp['normal'],rain['normal'])
rule4=ctrl.Rule(wind['veryhigh'] & temp['low'],rain['low'])
rule5=ctrl.Rule(wind['veryhigh'] & temp['verylow'],rain['verylow'])

rule6=ctrl.Rule(wind['high'] & temp['veryhigh'],rain['veryhigh'])
rule7=ctrl.Rule(wind['high'] & temp['high'],rain['high'])
rule8=ctrl.Rule(wind['high'] & temp['normal'],rain['normal'])
rule9=ctrl.Rule(wind['high'] & temp['low'],rain['low'])
rule10=ctrl.Rule(wind['high'] & temp['verylow'],rain['verylow'])

rule11=ctrl.Rule(wind['normal'] & temp['veryhigh'],rain['veryhigh'])
rule12=ctrl.Rule(wind['normal'] & temp['high'],rain['high'])
rule13=ctrl.Rule(wind['normal'] & temp['normal'],rain['normal'])
rule14=ctrl.Rule(wind['normal'] & temp['low'],rain['low'])
rule15=ctrl.Rule(wind['normal'] & temp['verylow'],rain['verylow'])

rule16=ctrl.Rule(wind['low'] & temp['veryhigh'],rain['veryhigh'])
rule17=ctrl.Rule(wind['low'] & temp['high'],rain['high'])
rule18=ctrl.Rule(wind['low'] & temp['normal'],rain['normal'])
rule19=ctrl.Rule(wind['low'] & temp['low'],rain['low'])
rule20=ctrl.Rule(wind['low'] & temp['verylow'],rain['verylow'])

rule21=ctrl.Rule(wind['verylow'] & temp['veryhigh'],rain['veryhigh'])
rule22=ctrl.Rule(wind['verylow'] & temp['high'],rain['high'])
rule23=ctrl.Rule(wind['verylow'] & temp['normal'],rain['normal'])
rule24=ctrl.Rule(wind['verylow'] & temp['low'],rain['low'])
rule25=ctrl.Rule(wind['verylow'] & temp['verylow'],rain['verylow'])

rainfall_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
rainfall=ctrl.ControlSystemSimulation(rainfall_ctrl)
rainfall.input['wind']=2.5
rainfall.input['temp']=30
rainfall.compute()
print(rainfall.output["rain"])
rain.view(sim=rainfall)
```

```python
#BMI
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

BMI = ctrl.Antecedent(np.arange(0,101,1),'BMI')
BMR = ctrl.Antecedent(np.arange(600,3501,1),'BMR') #mức độ hoạt động xét theo độ tuổi
Calo= ctrl.Consequent(np.arange(1200,3201,1),'Calo')

BMI['thin']=fuzz.trimf(BMI.universe,[0,17,18.5])
BMI['normal']=fuzz.trimf(BMI.universe,[18.5,20,25])
BMI['over']=fuzz.trimf(BMI.universe,[25,26,27])
BMI['obesity']=fuzz.trimf(BMI.universe,[27,100,100])

BMR['thin']=fuzz.trimf(BMR.universe,[600,600,1000])
BMR['normal']=fuzz.trimf(BMR.universe,[1000,1250,1500])
BMR['over']=fuzz.trimf(BMR.universe,[1500,2000,2500])
BMR['obesity']=fuzz.trimf(BMR.universe,[2500,3000,3500])

Calo['thin']=fuzz.trimf(Calo.universe,[1200,1600,2000])
Calo['normal']=fuzz.trimf(Calo.universe,[1600,2000,2600])
Calo['over']=fuzz.trimf(Calo.universe,[2400,2500,2600])
Calo['obesity']=fuzz.trimf(Calo.universe,[2500,2700,3200])


rule1=ctrl.Rule(BMI['thin'] & BMR['thin'],Calo['obesity'])
rule2=ctrl.Rule(BMI['thin'] & BMR['normal'],Calo['over'])
rule3=ctrl.Rule(BMI['thin'] & BMR['over'],Calo['normal'])
rule4=ctrl.Rule(BMI['thin'] & BMR['obesity'],Calo['normal'])

rule5=ctrl.Rule(BMI['normal'] & BMR['thin'],Calo['obesity'])
rule6=ctrl.Rule(BMI['normal'] & BMR['normal'],Calo['normal'])
rule7=ctrl.Rule(BMI['normal'] & BMR['over'],Calo['thin'])
rule8=ctrl.Rule(BMI['normal'] & BMR['obesity'],Calo['thin'])

rule9=ctrl.Rule(BMI['over'] & BMR['thin'],Calo['over'])
rule10=ctrl.Rule(BMI['over'] & BMR['normal'],Calo['normal'])
rule11=ctrl.Rule(BMI['over'] & BMR['over'],Calo['thin'])
rule12=ctrl.Rule(BMI['over'] & BMR['obesity'],Calo['thin'])

rule13=ctrl.Rule(BMI['obesity'] & BMR['thin'],Calo['normal'])
rule14=ctrl.Rule(BMI['obesity'] & BMR['normal'],Calo['thin'])
rule15=ctrl.Rule(BMI['obesity'] & BMR['over'],Calo['thin'])
rule16=ctrl.Rule(BMI['obesity'] & BMR['obesity'],Calo['thin'])

calorie_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16])
calorie=ctrl.ControlSystemSimulation(calorie_ctrl)
weight =100 #kg
height = 160 #cm
age = 20
sex = 'male'
BMI1 = weight/(height/100)**2
if sex =='male' :
  BMR1 = 10*weight+6.25*height-5*age+5
if sex =='female':
  BMR1 = 10*weight+6.25*height-5*age-161
calorie.input['BMI']=BMI1
calorie.input['BMR']=BMR1
calorie.compute()
print(calorie.output["Calo"])
Calo.view(sim=calorie)
```
