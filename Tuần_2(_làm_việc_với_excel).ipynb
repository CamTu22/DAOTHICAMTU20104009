```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(3)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.subplot(2,2,1)
plt.plot(x,-x**2,color='g')
plt.subplot(2,2,2)
plt.plot(x,-x**3,color='r')
plt.subplot(2,2,3)
plt.plot(x,-2*x,color='b')
plt.subplot(2,2,4)
plt.plot(x,-2**x,'o')
plt.show()
```

```python
# vẽ đường thẳng : plt.plot([xA,xB],[yA,yB],linewidth=5,color='r')
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,5],[2,7],linewidth=2,color='r')
plt.text(2,5,'CT') #viết chữ
plt.show()
```

```python
# vẽ mũi tên: plt.arrow(xA,yA,xB-xA,yB-yA,headlenghth=4,headwidth=2,color='g)
import matplotlib.pyplot as plt
plt.axis([-50,100,-50,100])
plt.arrow(30,30,0,-30, head_length=4, head_width=2, color='g')
plt.show
```

CÁC DẠNG ĐỒ THỊ KHÁC

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(4)
y = np.random.rand()
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.random.rand(3,4)
plt.bar(x+0,y[0],color='b',width=0.25)
plt.bar(x+0.25,y[1],color='g',width=0.25)
plt.bar(x+0.5,y[2],color='r',width=0.25)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(4)
y = np.random.rand(4)
plt.barh(x,y,color ='g',alpha=.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
n = 1000
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
size = 20
plt.scatter(x,y,s=size,c=colors,alpha=1)
plt.show() 
```
